```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.colors as mcolors
from matplotlib.dates import DateFormatter

sns.set()
%config InlineBackend.figure_format = 'retina'

pd.set_option("display.precision", 2)
mpl.rcParams['timezone'] = 'UTC'
```

```python
DATA_URL="../data/raw_sales.csv"
ACCURACIES_URL="../results/rf-results-sales.csv"
```

```python
#df = pd.read_csv(DATA_URL);
df_results = pd.read_csv(ACCURACIES_URL);
```

```python
numerical = list(set(df_results.columns) - set(["criterion"]))

# Calculate and plot
corr_matrix = df_results[numerical].corr()
sns.heatmap(corr_matrix)
```

```python
numerical = list(set(df_results.columns) - set(["accuracy"]))
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
for idx, feat in enumerate(numerical):
    ax = axes[int(idx / 2), idx % 2]
    sns.boxplot(x=feat, y="accuracy", data=df_results, ax=ax)
    ax.set_xlabel("")
    ax.set_ylabel(feat)
fig.tight_layout();
```

```python
#Same as above, just for num_lags = 10
df = df_results[df_results['num_lags'] == 10]
numerical = list(set(df.columns) - set(["accuracy"]))
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
for idx, feat in enumerate(numerical):
    ax = axes[int(idx / 2), idx % 2]
    sns.boxplot(x=feat, y="accuracy", data=df, ax=ax)
    ax.set_xlabel("")
    ax.set_ylabel(feat)
fig.tight_layout();
```
